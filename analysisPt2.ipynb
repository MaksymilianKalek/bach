```python
import pandas as pd
```

```python
complete_data = pd.read_csv('complete_data.csv')
complete_data
```

```python
only_necessary_data = complete_data.drop(complete_data.columns[0], axis=1)
only_necessary_data = only_necessary_data.drop(only_necessary_data.columns[0], axis=1)
only_necessary_data
```

```python
renamed_columns = only_necessary_data.rename(columns={'grid': 'startingGridPosition', 'code': 'driverCode', 'name_x': 'raceName', 'diff': 'positionDiff', 'name_y': 'teamName', 'name': 'circuitName'})
sorted_data = renamed_columns.sort_values(by=['raceId', 'position'])
sorted_data
```

```python
sorted_data.to_csv('sorted_data.csv')
```

```python
sorted_data = pd.read_csv('sorted_data.csv', index_col=0)
sorted_data
```

```python
sorted_data.count(axis = 0)
```

```python
most_important = sorted_data[['raceId', 'driverCode', 'teamName', 'lap', 'tireChange', 'startingGridPosition', 'position', 'positionDiff', 'circuitName', 'raceName', 'year', 'AirTemp', 'Humidity', 'TrackTemp', 'WindSpeed', 'WasRainingDuringRaceAnyTime']]
most_important.fillna({'tireChange':'UNKNOWN'}, inplace=True)
most_important["lapTireChange"] = most_important['lap'].astype(str) + ': ' + most_important["tireChange"]
most_important
```

```python
grouped = most_important.groupby(['raceId', 'driverCode', 'teamName', 'startingGridPosition', 'position', 'positionDiff', 'circuitName', 'raceName', 'year', 'AirTemp', 'Humidity', 'TrackTemp', 'WindSpeed', 'WasRainingDuringRaceAnyTime'])['lapTireChange'].apply(', '.join).reset_index()
# grouped.drop([''])
grouped
```
